```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts

```

```python
mart = pd.read_excel("supermarket_sales_data.xlsx")
mart = mart[['Gender','Payment','Unit price','Quantity','Total','gross income']]
mart.head(10)
```

```python
sns.histplot(data=mart,x='Total')
```

```python
sns.histplot(data=mart,x='Total',binwidth=150)
```

```python
# Adjust number of bins and the X-Axis interval
sns.histplot(data=mart, kde=False, bins=20)
plt.title("Histogram with Custom Number of Bins")
plt.show()

```

```python

# Changing the X-axis interval (ticks)
plt.figure(figsize=(10,6))
sns.histplot(data=mart, kde=False, bins=20)
plt.xticks(np.arange(-4, 4, 1))  
plt.title("Histogram with Custom X-Axis Interval")
plt.show()
```

```python
sns.histplot(data=mart,x='Total',kde=True)
```

```python
sns.histplot(data=mart,x='Total',hue='Payment',multiple='stack')
```

```python
sns.histplot(data=mart,x='Total',hue='Payment')
```

```python
sns.histplot(data=mart,x='Total',hue='Payment',element='poly')
```

```python
sns.histplot(data=mart,x='Total',hue='Payment',element='step')
```

```python
sns.histplot(data=mart,x='Total',element='poly')
```

```python
sns.histplot(data=mart,x='Payment',element='bars')
```

```python
sns.histplot(data=mart,x='Total',stat='probability')
```

```python
sns.histplot(data=mart,x='Total',y='gross income')
```

```python


sns.histplot(data=mart,x='Total',y='gross income',hue='Gender')
plt.title("Auto Shrinking with Categorical Hue")
plt.show()
```

```python
# KDE gives you a smoothed estimate of the distribution of your data.
# It helps visualize the shape of the data, showing where the data points are more concentrated and where there are gaps.
sns.histplot(data=mart, kde=True, stat="density")
plt.title("Histogram with Density")
plt.show()
```

```python

# Use a cumulative sum
sns.histplot(data=mart, kde=True, cumulative=True)
plt.title("Cumulative Histogram")
plt.show()
```

```python
sns.histplot(data=mart,x='Payment',shrink=0.7)
 
```
