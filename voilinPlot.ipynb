```python
# # A Violin Plot is a combination of a box plot and a kernel density plot. 
#It’s a way to show the distribution of the data, giving more insight into the underlying distribution of the variable.

# Here’s what a violin plot consists of:

# Box Plot: The box part of the plot shows the median, quartiles, and outliers, similar to a regular box plot.
# Density Plot: The "violin" shape represents the kernel density estimation (KDE),
#which is a smooth curve that shows the distribution of the data.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart = pd.read_excel("supermarket_sales_data.xlsx")
mart.head()
```

```python
sns.violinplot(y="Total",data=mart)
#right side is an image of left side (its nothing but normal distributiion)shape of voilin represent distribution of data
```

```python
sns.violinplot(x='Payment',y='Total',hue='Gender',data=mart)
```

```python
sns.violinplot(x='Payment',y='Total',hue='Gender',split=True,data=mart)
```

```python
#smoothness of voilin by bw
sns.violinplot(x='Payment', y='Total', split=True, bw_method=0.5, data=mart)

# Smaller BW: The curve will be more jagged (showing more detail).
# Larger BW: The curve will be smoother (less detail).
```

```python
#observing by lines ,in this case there are alot of observations (not usually required)
sns.violinplot(x='Payment',y='Total',inner="stick",data=mart)
```

```python
#extreme values has been removed by cut 
sns.violinplot(x='Payment',y='Total',hue='Gender',data=mart,split=True,inner='quartile',bw_method=.2,cut=0)
```
