```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

# Load Titanic dataset from a local file
titanic = pd.read_csv("titanic (1).csv")

# Display the first 5 rows
print(titanic.head())

```

```python
# 1. Basic Bar Plot
plt.figure(figsize=(15,5))
sns.barplot(x='class', y='fare', data=titanic)
plt.title("Average Fare by Class")
plt.show()
```

```python
# 2. Add HUE to the barplot
plt.figure(figsize=(10,5))
sns.barplot(x='class', y='fare', hue='sex', data=titanic)
plt.title("Average Fare by Class and Gender")
plt.show()
```

```python
sns.barplot(x='fare', y='class', data=titanic)
plt.title("Average Fare by Class (Horizontal)")
plt.show()
```

```python
# 4. Plot bars in a given order
import seaborn as sns
sns.barplot(x='class', y='fare', hue='sex', data=titanic, order=['First', 'Second', 'Third'], hue_order=['male', 'female'])
plt.title("Average Fare by Class and Gender (Ordered)")
plt.show()
```

```python
# 5. Add CAP on the ERROR BAR

sns.barplot(x='class', y='fare', hue='sex', data=titanic, capsize=0.3)
plt.title("Average Fare by Class and Gender with Capsize")
plt.show()
```

```python
# 5. Add CAP on the ERROR BAR
plt.figure(figsize=(15,5))
sns.barplot(x='class', y='fare', hue='sex', data=titanic, capsize=0.2)
plt.title("Average Fare by Class and Gender with Capsize")
plt.show()






```

```python

# 6. Remove Error Bar
plt.figure(figsize=(15,5))
sns.barplot(x='class', y='fare', hue='sex', data=titanic, ci=None)
plt.title("Average Fare by Class and Gender (No Error Bars)")
plt.show()
```

```python
# 7. Change Bar Colors using COLOR attribute
plt.figure(figsize=(15,5))
sns.barplot(x='class', y='fare', data=titanic, color='red')
plt.title("Average Fare by Class (Red Bars)")
plt.show()
```

```python
# 8. Change Colors using PALETTE attribute
plt.figure(figsize=(15,5))
sns.barplot(x='class', y='fare', hue='sex', data=titanic, palette="pastel")
plt.title("Average Fare by Class and Gender (Palette)")
plt.show()
```

```python
# 9. Using SATURATION parameter
plt.figure(figsize=(15,5))
sns.barplot(x='class', y='fare', hue='sex', data=titanic, palette='cividis', saturation=0.1)
plt.title("Average Fare by Class and Gender (Low Saturation)")
plt.show()
```

```python
# 10. Change Default Aggregation Method using ESTIMATOR parameter
plt.figure(figsize=(15,5))
sns.barplot(x='class', y='fare', hue='sex', data=titanic, estimator=np.median)
plt.title("Median Fare by Class and Gender")
plt.show()
```
