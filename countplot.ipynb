```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Create a DataFrame
data = {
    'Person_ID': range(1, 11),
    'Age_Group': ['18-24', '25-34', '35-44', '18-24', '45-54', '25-34', '35-44', '45-54', '18-24', '25-34'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Product_Satisfaction': ['Satisfied', 'Neutral', 'Unsatisfied', 'Satisfied', 'Neutral', 'Satisfied', 'Unsatisfied', 'Satisfied', 'Neutral', 'Unsatisfied']
}

df = pd.DataFrame(data)

# Question 1: How many respondents in each age group
sns.countplot(x='Age_Group', data=df)
plt.title('Number of Respondents in Each Age Group')
plt.show()
```

```python
# Question 2: Product satisfaction by age group
sns.countplot(x='Age_Group', hue='Product_Satisfaction', data=df)
plt.title('Product Satisfaction by Age Group')
plt.show()


```

```python
# Question 3: Gender distribution
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.show()
```

```python
# Question 4: Gender vs Product Satisfaction
sns.countplot(x='Gender', hue='Product_Satisfaction', data=df)
plt.title('Gender vs Product Satisfaction')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Gender', hue='Product_Satisfaction', data=df)
plt.title('Product Satisfaction by Age Group')
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(x="Product_Satisfaction",hue="Gender",data=df,palette='PuBuGn_r')
```
